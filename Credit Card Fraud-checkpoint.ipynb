```python
#separating the data for analysis
legit_transactions= credit_card_data[credit_card_data.Class == 0]
fraudulent_transactions= credit_card_data[credit_card_data.Class == 1]
```

UNDER SAMPLING

It is done as a means of dealing with the imbalanced data. A sample dataset containing similar distribution of legit and fraudulent transactions is created.

```python
legit_sample=legit_transactions.sample(n=492)
legit_sample
```

```python
#Concatenating two dataframes
new_dataset= pd.concat([legit_sample, fraudulent_transactions], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

DATA SPLITTING

```python
#splitting the data into features and targets
X= new_dataset.drop(columns= 'Class', axis=1)
Y= new_dataset['Class']
print(X)
```

```python
print(Y)
```

X represents the features while Y represents the target

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting the new dataset into training and testing data
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING AND EVALUATION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Logistic Regression is used in binary classification as is the case in our dataset.

```python
model= LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#Accuracy score
#Accuracy on training data
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
print('Accuracy on training data: ', training_data_accuracy)
```

```python
#Accuracy on testing data
X_test_prediction= model.predict(X_test)
testing_data_accuracy= accuracy_score(X_test_prediction, Y_test)
print('Accuracy on testing data: ', testing_data_accuracy)
```

Accuracy on training and trainig data is nearly equal hence the model has performed well.
